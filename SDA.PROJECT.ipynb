## Aggregate data per user

[Now, as the data is clean, aggregate data per user per period in order to have just one record per user per period. It should ease the further analysis a lot.]

```python
# Calculate the number of calls made by each user per month. Save the result.
calls_per_month = (
    calls.groupby(['user_id', 'month'], as_index=False)
         .agg(calls_count=('id', 'count'))
)
```

```python
# Calculate the amount of minutes spent by each user per month. Save the result.
minutes_per_month = (
    calls.groupby(['user_id', 'month'], as_index=False)
         .agg(total_minutes=('call_min_ceil', 'sum'))
)
```

```python
# Calculate the number of messages sent by each user per month. Save the result.
messages_per_month = (
    messages.groupby(['user_id', 'month'], as_index=False)
            .agg(messages_count=('id', 'count'))
)
```

```python
# Calculate the volume of internet traffic used by each user per month. Save the result.
internet_per_month = (
    internet.groupby(['user_id', 'month'], as_index=False)
            .agg(total_mb=('mb_used', 'sum'))
)
```

[Put the aggregate data together into one DataFrame so that one record in it would represent what an unique user consumed in a given month.]

```python
# Merge the data for calls, minutes, messages, internet based on user_id and month
# Put the aggregate data together into one DataFrame so that one record in it would represent
# what an unique user consumed in a given month.
user_monthly_data = calls_per_month.merge(minutes_per_month, on=['user_id', 'month'], how='outer')
user_monthly_data = user_monthly_data.merge(messages_per_month, on=['user_id', 'month'], how='outer')
user_monthly_data = user_monthly_data.merge(internet_per_month, on=['user_id', 'month'], how='outer')

# Fill NaN with zeros (no usage in that category for the month)
user_monthly_data[['calls_count', 'total_minutes', 'messages_count', 'total_mb']] = \
    user_monthly_data[['calls_count', 'total_minutes', 'messages_count', 'total_mb']].fillna(0)
```

```python
# Add the plan information
user_monthly_data = user_monthly_data.merge(users[['user_id', 'plan']], on='user_id', how='left')
user_monthly_data = user_monthly_data.merge(plans, left_on='plan', right_on='plan_name', how='left')
```

[Calculate the monthly revenue from each user (subtract the free package limit from the total number of calls, text messages, and data; multiply the result by the calling plan value; add the monthly charge depending on the calling plan). N.B. This might not be as trivial as just a couple of lines given the plan conditions! So, it's okay to spend some time on it.]

```python
# Calculate the monthly revenue for each user
def calculate_revenue(row):
    # Start with base monthly fee
    revenue = row['usd_monthly_pay']
    
    # Minutes overage
    if row['total_minutes'] > row['minutes_included']:
        revenue += (row['total_minutes'] - row['minutes_included']) * row['usd_per_minute']
    
    # Messages overage
    if row['messages_count'] > row['messages_included']:
        revenue += (row['messages_count'] - row['messages_included']) * row['usd_per_message']
    
    # Internet overage: convert MB to GB, round up
    gb_used = np.ceil(row['total_mb'] / 1024)
    if gb_used > (row['mb_per_month_included'] / 1024):
        revenue += (gb_used - (row['mb_per_month_included'] / 1024)) * row['usd_per_gb']
    
    return revenue

# Calculate the monthly revenue for each user
user_monthly_data['monthly_revenue'] = user_monthly_data.apply(calculate_revenue, axis=1)

display(user_monthly_data.head())
```

```python
# Compare average duration of calls per each plan per each distinct month. Plot a bar plat to visualize it.
# Build a calls+plan frame (uses per-call rounded minutes)
calls_with_plan = calls.merge(users[['user_id','plan']], on='user_id', how='left')

avg_call_len = (
    calls_with_plan.groupby(['plan','month'], as_index=False)
                   .agg(avg_call_minutes=('call_min_ceil','mean'),
                        call_count=('id','count'))
)
# Pivot for bar plot
avg_call_pivot = avg_call_len.pivot(index='month', columns='plan', values='avg_call_minutes').sort_index()
avg_call_pivot.index = avg_call_pivot.index.astype(str)  # cleaner x-axis labels

ax = avg_call_pivot.plot(kind='bar', figsize=(12,5))
ax.set_title('Average Call Duration by Plan and Month')
ax.set_xlabel('Month (2018)')
ax.set_ylabel('Average call duration (minutes)')
ax.legend(title='Plan')
plt.tight_layout()
plt.show()
```

## Test statistical hypotheses

[Test the hypothesis that the average revenue from users of the Ultimate and Surf calling plans differs.]

[Formulate the null and the alternative hypotheses, choose the statistical test, decide on the alpha value.]

```python
# 1.15 Statistical Hypotheses — keep equality in H0 and concise decisions

from scipy import stats

alpha = 0.05

# Test 1: Ultimate vs Surf revenue
# H0: μ_Ultimate = μ_Surf
# H1: μ_Ultimate ≠ μ_Surf

surf_rev = user_monthly_data.loc[user_monthly_data['plan']=='surf', 'monthly_revenue']
ult_rev  = user_monthly_data.loc[user_monthly_data['plan']=='ultimate', 'monthly_revenue']

t_stat_1, p_val_1 = stats.ttest_ind(ult_rev, surf_rev, equal_var=False, nan_policy='omit')
print("Test 1 — Plan revenue difference (Welch t-test)")
print(f"H0: μ_Ultimate = μ_Surf | t = {t_stat_1:.3f}, p = {p_val_1:.4f}, alpha = {alpha}")
print("Decision:", "Reject H0 (evidence of difference)." if p_val_1 < alpha else "Fail to reject H0 (no evidence of difference).")
```

Hypothesis 1: Average revenue differs between Ultimate and Surf.
H₀: μ_Ultimate = μ_Surf
H₁: μ_Ultimate ≠ μ_Surf

Test: Welch’s t‑test (α = 0.05).

Result: t = 7.952, p = < 0.0001 → Reject H₀. 
Average revenue does differ between plans, with Ultimate higher on average (see 1.14). 

```python
# Test 2: NY–NJ vs. Others
# (Assumes you’ve already created users['is_ny_nj_area'] earlier; if not, uncomment next line)
# users['is_ny_nj_area'] = users['city'].str.contains(r'\bNY-NJ\b', regex=True, na=False)

rev_geo = user_monthly_data.merge(users[['user_id','is_ny_nj_area']], on='user_id', how='left')
ny_nj_rev = rev_geo.loc[rev_geo['is_ny_nj_area']==True,  'monthly_revenue']
other_rev = rev_geo.loc[rev_geo['is_ny_nj_area']!=True, 'monthly_revenue']

t_stat_2, p_val_2 = stats.ttest_ind(ny_nj_rev, other_rev, equal_var=False, nan_policy='omit')
print("\nTest 2 — NY–NJ vs Other regions (Welch t-test)")
print(f"H0: μ_NY–NJ = μ_Other | t = {t_stat_2:.3f}, p = {p_val_2:.4f}, alpha = {alpha}")
print("Decision:", "Reject H0 (evidence of difference)." if p_val_2 < alpha else "Fail to reject H0 (no evidence of difference).")
```

Hypothesis 2: Average revenue in the NY–NJ area differs from other regions.

Test: Welch’s t‑test (α = 0.05).
Result: t = −2.131, p = 0.0335 → Reject H₀.
